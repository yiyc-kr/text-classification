```python
import pandas as pd
df = pd.read_csv('data/list.csv', names=["class_id", "name"], dtype={"class_id": str, "name": str}, 
                 skiprows=1, header=None)
df.head()
```

```python
import re
df.name = df.name.replace(r'[^a-zA-Z\- ]+', ' ', regex=True)\
                    .replace(r'\B[\-]\B|\B[\-]|[\-]\B', '', regex=True)\
                    .replace(r'\s[a-zA-Z]\s|\s[a-zA-Z]$|^[a-zA-Z]\s', '', regex=True)\
                    .replace(r'\s{2,}', ' ', regex=True)\
                    .str.upper()
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df = df.drop(df[df.name == ''].index).drop(df[df.name == ' '].index)
df.info()
```
